```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv('job_postings.csv') #declare a dataframe
df
```

```python
df['description'] =  df.description.str.encode('utf-8').str.decode('ascii', 'ignore') #ignoring the ascii in the description
```

```python
df.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull()
```

```python
df['company_id'] = df['company_id'].fillna('NA')
```

```python
df['description'] = df['description'].fillna('No description')
```

```python
df['max_salary'] = df['max_salary'].fillna('NA')
```

```python
df['med_salary'] = df['med_salary'].fillna('NA')
```

```python
df['min_salary'] = df['min_salary'].fillna('NA')
```

```python
df['pay_period'] = df['pay_period'].fillna('No pay period available')
```

```python
df['applies'] = df['applies'].fillna('No applies')
```

```python
df['remote_allowed'] = df['remote_allowed'].fillna('Not allowed')
```

```python
df['views'] = df['views'].fillna('views not available')
```

```python
df['application_url'] = df['application_url'].fillna('No application url available')
```

```python
df['closed_time'] = df['closed_time'].fillna('Not mentioned')
```

```python
df['formatted_experience_level'] = df['formatted_experience_level'].fillna('Not available')
```

```python
df['skills_desc'] = df['skills_desc'].fillna('No skills mentioned')
```

```python
df['posting_domain'] = df['posting_domain'].fillna('No domain')
```

```python
df['currency'] = df['currency'].fillna('NA')
```

```python
df['compensation_type'] = df['compensation_type'].fillna('Not mentioned')
```

```python
df.to_csv("job_postings_cleaned.csv", encoding="utf-8")
```
